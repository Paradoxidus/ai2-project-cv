## Training

```python
from ultralytics import YOLO

# Initialize the YOLO model
# ⚠️ Use the appropriate checkpoint:
# - yolo12n.pt       → for object detection
# - yolo11n-seg.pt   → for object segmentation
model = YOLO('/content/drive/MyDrive/Proj_YOLOv12/yolov12/yolo11n-seg.pt')

# Train the model
results = model.train(
    data='/content/drive/MyDrive/Proj_YOLOv12/braintumor_yolov12v5/data.yaml',  # dataset config file
    epochs=1000,       # maximum epochs (early stopping may end sooner)
    patience=15,       # stop if no improvement for 15 epochs
    batch=8,           # batch size
    imgsz=640,         # input image size
    scale=0.5,         # data augmentation scaling
    mosaic=1.0,        # mosaic augmentation probability
    mixup=0.0,         # mixup probability
    copy_paste=0.1,    # copy-paste augmentation probability
    close_mosaic=10,   # disable mosaic augmentation in the last 10 epochs
    device="0",        # set to "0" for GPU, "cpu" for CPU
    save=True          # save both best.pt and last.pt
)
```

## Testing

```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Load trained YOLOv12 segmentation model weights
model = YOLO("/content/drive/MyDrive/Proj_YOLOv12/yolov12/runs/segment/train4/weights/best.pt")

# Test image path
img_path = "/content/drive/MyDrive/Proj_YOLOv12/braintumor_yolov12v5/test/images/enh_Tr-pi_1450_jpg.rf.99a73245080d1e77b9be78f97d32885c.jpg"

# Run inference
results = model(img_path)

# Process and visualize results
for r in results:
    # Load the original image (full resolution)
    img = cv2.imread(img_path)

    # Overlay segmentation masks
    if r.masks is not None:
        masks = r.masks.data.cpu().numpy()
        for mask in masks:
            mask = mask.astype(np.uint8) * 255
            colored_mask = np.zeros_like(img)
            colored_mask[:, :, 2] = mask  # Red overlay for segmentation
            img = cv2.addWeighted(img, 1.0, colored_mask, 0.5, 0)

    # Draw bounding boxes, labels, and confidence scores
    boxes = r.boxes.xyxy.cpu().numpy()
    scores = r.boxes.conf.cpu().numpy()       # confidence scores
    class_ids = r.boxes.cls.cpu().numpy().astype(int)  # class IDs
    names = model.names                       # class name mapping

    for box, score, cls_id in zip(boxes, scores, class_ids):
        x1, y1, x2, y2 = map(int, box)
        label = f"{names[cls_id]} {score:.2f}"  # e.g. "tilapia 0.93"

        # Draw rectangle
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Put label + score above the box
        cv2.putText(img, label, (x1, max(y1 - 10, 0)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2,
                    lineType=cv2.LINE_AA)

    # Display image at original resolution
    h, w = img.shape[:2]
    dpi = 300
    fig_w, fig_h = w / dpi, h / dpi
    fig, ax = plt.subplots(figsize=(fig_w, fig_h), dpi=dpi)

    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), interpolation='nearest')
    ax.axis("off")
    plt.show()
```

## **2. Research-Style Methodology and Discussion**

### **2.1 Methodology**

Dataset Preprocessing:
Images are resized to 640×640, normalized, and augmented using YOLO’s native techniques (mosaic, mixup, copy-paste).

Model Configuration:
YOLOv12 uses an attention-optimized backbone and transformer-based neck for better feature fusion. Training applies multi-loss optimization combining box, class, segmentation, and distribution focal losses.

Training Environment:
Training and validation were performed on Google Colab GPUs with CUDA acceleration. The optimizer (SGD / AdamW) used dynamic learning-rate scheduling and early stopping after 15 stagnant epochs.

Evaluation Strategy:
Model performance was evaluated using mAP@0.5, mAP@0.5–0.95, precision, and recall on the validation set.

---

### **2.2 Results and Discussion**

Detection (Box) Performance

- mAP@0.5 = 0.8408

- mAP@0.5–0.95 = 0.5539

- Precision (avg) = 0.82

- Recall (avg) = 0.75

- Validation Box Loss = 1.2364

- Validation Class Loss = 0.8686

- Segmentation Loss = 1.8896

Interpretation:

- mAP@0.5 = 0.8408 → indicates high detection accuracy at a moderate IoU threshold.
- mAP@0.5–0.95 = 0.5539 → reflects consistent performance across stricter IoU thresholds.
- Precision ≈ 0.82 / Recall ≈ 0.75 → balanced detection performance with limited false positives and false negatives.
- Validation losses show stable learning with minimal overfitting.

---

Segmentation (Mask) Performance:

- mAP@0.5 = 0.8517
- mAP@0.5–0.95 = 0.5436
- Precision (avg) = 0.85
- Recall (avg) = 0.76
- Validation Mask Loss = 1.8516

Interpretation:

The Mask Precision–Confidence Curve illustrates how precision varies across confidence thresholds for Glioma, Meningioma, and Pituitary tumor classes.
- Meningioma maintained the highest precision overall, consistently staying above 0.85 across most confidence levels.

- Glioma and Pituitary exhibited stable trends, with precision notably improving beyond a 0.5 confidence threshold, reflecting strong confidence calibration.

- The overall precision peaked at 1.00 near a confidence level of 0.86, indicating exceptional reliability when high-confidence predictions are considered.

- These outcomes confirm that YOLOv12’s segmentation head effectively captures and distinguishes tumor boundaries, achieving robust generalization across all three tumor types with balanced precision and recall.

---

Overall Discussion:
The YOLOv12 architecture proved effective for both multi-class detection and segmentation of brain tumors. Its transformer-based backbone enhances spatial feature representation, enabling precise tumor localization and boundary definition.

A slight gap between training and validation losses suggests mild overfitting, which could be mitigated by:
- Increasing augmentation intensity (stronger mosaic or mixup)
- Applying dropout or weight decay
- Expanding the dataset with more MRI variations

Overall, the trained YOLOv12 model provides a robust baseline for automated brain tumor localization and classification. It demonstrates promising generalization, accuracy, and inference efficiency, making it suitable for real-world medical imaging research.
